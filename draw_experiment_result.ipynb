```python
import wandb
import matplotlib.pyplot as plt

# Initialize wandb
api = wandb.Api(timeout=60)

# List of run ids
run_ids = ["hantao-zhong/higgs-graphs/fta13x39", "hantao-zhong/higgs-graphs/36uklm5e", "hantao-zhong/higgs-graphs/4mkb3h34"]
run_names = ["Dual-Branch GNN", "Heterogeneous Graph from Embedding Space", "Feynman Diagram Graph"]

plt.figure(figsize=(12, 6))

# Loop over the run ids
for i, run_id in enumerate(run_ids):
    # Access the run
    run = api.run(run_id)

    # Get the history
    history = run.history()

    # Extract the train_loss and val_loss
    train_loss = history["train_loss"]
    val_loss = history["val_loss"]

    # Plot the train_loss and val_loss
    run_name = run_names[i]
    color = 'C' + str(i)  # Use matplotlib's default color cycle
    plt.plot(train_loss, label=f'Train Loss - {run_name}', color=color)
    plt.plot(val_loss, label=f'Validation Loss - {run_name}', linestyle=':', color=color)

plt.title('Train and Validation Loss for Graph Classification Models')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
