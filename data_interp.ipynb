```python
import pandas as pd
higgs_data = pd.read_csv('./data/HIGGS.csv.gz', compression='gzip', header=None, nrows=80000)
higgs_data.columns = ['class_label',
                      'lepton_pT', 'lepton_eta', 'lepton_phi',
                      'missing_energy_magnitude', 'missing_energy_phi',
                      'jet_1_pt', 'jet_1_eta', 'jet_1_phi', 'jet_1_b-tag',
                      'jet_2_pt', 'jet_2_eta', 'jet_2_phi', 'jet_2_b-tag',
                      'jet_3_pt', 'jet_3_eta', 'jet_3_phi', 'jet_3_b-tag',
                      'jet_4_pt', 'jet_4_eta', 'jet_4_phi', 'jet_4_b-tag',
                      'm_jj', 'm_jjj', 'm_lv', 'm_jlv', 'm_bb', 'm_wbb', 'm_wwbb']
```

```python
higgs_data.describe().T
```

```python
def normalize(df):
    return (df - df.min()) / (df.max() - df.min())

normalized_higgs_data = normalize(higgs_data)
normalized_higgs_data.describe().T
```

```python
x = higgs_data.drop('class_label', axis=1)
y = higgs_data['class_label']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Assuming X_train and y_train are your features and labels
gb_clf = GradientBoostingClassifier(random_state=0)
gb_clf.fit(x_train, y_train)
```

```python
gb_clf.score(x_test, y_test)
```

```python
# calculate roc auc score
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, gb_clf.predict(x_test))
```

```python
from xgboost import XGBClassifier, plot_tree

xgb_clf = XGBClassifier()
xgb_clf.fit(x_train, y_train)

xgb_clf.score(x_test, y_test)
```

```python
xgb_clf_pred = xgb_clf.predict(x_test)
roc_auc_score(y_test, xgb_clf_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, xgb_clf_pred))
print(classification_report(y_test, xgb_clf_pred))
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(100, 100))
plot_tree(xgb_clf, ax=ax)
plt.show()
```
